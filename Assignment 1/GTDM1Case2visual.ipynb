```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# plot the feasible region
d = np.linspace(-20000,45000,20000)
x,y = np.meshgrid(d,d)
plt.imshow( ((y>=-19/4*x+775) & (y>=-3/41*x+4340/41) & (y<=-7/17*x+237770/17) & (y>=0) & (x>=0)).astype(int) , 
                extent=(x.min(),x.max(),y.min(),y.max()),origin="lower", cmap="Greys", alpha = 0.3);


# plot the lines defining the constraints
x = np.linspace(-20000, 45000, 20000)
#
y1 = -19/4*x+775
# 
y2 = -3/41*x+4340/41
# 
y3 = -7/17*x+237770/17



# Make plot
#plt.plot(x, 2*np.ones_like(y1))
plt.plot(x, y1, label=r'$x_a \geq 0$')
plt.plot(x, y2, label=r'$x_b \geq 0$')
plt.plot(x, y3, label=r'$x_c \geq 0$')

plt.xlim(-20000,45000)
plt.ylim(-10000,20000)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel(r'$a_2$, i.e. electricity')
plt.ylabel(r'$a_3$, i.e. packaging')



plt.savefig('GTDMplane.png', bbox_inches='tight')
```
